```python
import os
import re
import json


out = 'maps'
```

Convert the age maps

```python
dir = 'age_maps'
ll = os.listdir( dir )
reg = r'Month\s(\d*)\sof\s(\d{4})-(\d+)\sfor\s(.*).png'
# reg = r'Month\s(\d*)\sof\s(\d{4})-(\d+)\sfor\s(.*)'
features = set()
files = []

# os.mkdir( f'{dir}/maps' )
for s in ll:
  match = re.match( reg, s )
  if match:
    month = match.group( 1 )
    year = match.group( 2 )
    feat = match.group( 4 )
    if feat not in features:
      features.add( feat )
    files.append( [ s, month, year, feat ] )
# print( features )

cleanFeatures = set()
while features:
  s = features.pop()
  if not re.match( r'.*\(\d\)', s ):
    cleanFeatures.add( s )
features = cleanFeatures
print( list( features ) )
# print( json.dumps( features, indent=4 ) )

#move files 
for f in files:
  print( f'{dir}/{f[0]} to maps/{ f[1] }-{ f[2] }-{ f[3] }.png' )
  try:
    os.rename( f'{dir}/{f[0]}', f'{dir}/maps/{ f[1] }-{ f[2] }-{ f[3] }.png')
  except OSError as e:
    print( e )
```

Convert the race/ethnic maps

```python
dir = 're_maps'
ll = os.listdir( dir )
reg = r'(.*?)_(\d*).*(\d{4}).*.png'
features = set()
files = []

# os.mkdir( f'{out}/re' )
for s in ll:
  match = re.match( reg, s )
  if match:
    month = match.group( 2 )
    year = match.group( 3 )
    feat = match.group( 1 )
    if feat not in features:
      features.add( feat )
    files.append( [ s, month, year, feat ] )
# print( features )

# cleanFeatures = set()
# while features:
#   s = features.pop()
#   if not re.match( r'.*\(\d\)', s ):
#     cleanFeatures.add( s )
# features = cleanFeatures
print( list( features ) )
# print( json.dumps( features, indent=4 ) )

#move files 
for f in files:
  print( f'{dir}/{f[0]} to {out}/re/{ f[1] }-{ f[2] }-{ f[3] }.png' )
  try:
    os.rename( f'{dir}/{f[0]}', f'{out}/re/{ f[1] }-{ f[2] }-{ f[3] }.png')
  except OSError as e:
    print( e )
```
